```python
import numpy as np
import pandas as pd


df_data = pd.read_table('../arff/new_data.tsv')
```

```python
df_data.head()
```

```python
df_data.replace('.', '?', inplace=True)
```

```python
df_data.dtypes
```

```python
df_data['Label'].replace('CASE', 1, inplace = True)
df_data['Label'].replace('CTRL', 0, inplace = True)

df_data['GT'].replace('1/1', 1, inplace = True)
df_data['GT'].replace('0/1', 0, inplace = True)


df_data = pd.get_dummies(df_data, columns=['Inheritence_mode'])
```

```python
df_data['CADD_impute'] = [1 if ele  == '?' else 0 for ele in df_data['CADD']]
df_data['GWAVA_impute'] = [1 if ele  == '?' else 0 for ele in df_data['GWAVA']]
df_data['Similarity_score_impute'] = [1 if ele  == '?' else 0 for ele in df_data['Similarity_score']]
df_data['DANN_impute'] = [1 if ele  == '?' else 0 for ele in df_data['DANN']]

df_data.replace('?', np.nan, inplace=True)


df_data['CADD'] = pd.to_numeric(df_data['CADD'], errors='coerce')
df_data['DANN'] = pd.to_numeric(df_data['DANN'], errors='coerce')
df_data['GWAVA'] = pd.to_numeric(df_data['GWAVA'], errors='coerce')
df_data['Similarity_score'] = pd.to_numeric(df_data['Similarity_score'], errors='coerce')

    
df_data.fillna(df_data.mean(), inplace = True)
```

```python
df_data.to_csv('PVP_new_data.csv', index = None)
```
